```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)
```

```python
plt.figure(1)
plt.subplot(121)
plt.plot(X,C, 'r+')
plt.title('subplot1')
plt.subplot(122)
plt.plot(X,C)
plt.title('subplot2')

plt.figure(2)
plt.plot(X,C)
plt.figure(2) # figurefigurefigurefigure
plt.plot(X,S)
plt.figure(3)
plt.plot(X,S)
plt.figure(2)
plt.plot([1,2,3,4])
plt.figure(1)  #figure
plt.subplot(121)
plt.ylabel('cos')
plt.subplot(122)
plt.plot(X,S,linewidth=2.9)
plt.xlabel('x')
plt.figure(3)
plt.text(4,0,'x-sinx') #xy
plt.axis([-10,10,-2,2]) #xy
plt.grid(True) #

```
