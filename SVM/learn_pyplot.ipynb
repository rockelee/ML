```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-np.pi, np.pi, 256,endpoint=True)
C,S = np.cos(X), np.sin(X)
```

```python
plt.figure(1)
plt.subplot(121)
plt.plot(X,C, 'r+')
plt.title('subplot1')
plt.subplot(122)
plt.plot(X,C)
plt.title('subplot2')

plt.figure(2)
plt.plot(X,C)
plt.figure(2) # 这里没有增加figure的编号，那么就会在之前的figure区域中画图，figure的编号是指定在某个figure区域中进行接下来的操作
plt.plot(X,S)
plt.figure(3)
plt.plot(X,S)
plt.figure(2)
plt.plot([1,2,3,4])
plt.figure(1)  #这里指定在第一个figure区域中操作
plt.subplot(121)
plt.ylabel('cos')
plt.subplot(122)
plt.plot(X,S,linewidth=2.9)
plt.xlabel('x')
plt.figure(3)
plt.text(4,0,'x-sinx') #在图像中插入文本，前面分别是x和y的坐标（相对于坐标轴的坐标）
plt.axis([-10,10,-2,2]) #这里设置x和y坐标轴的范围
plt.grid(True) #在坐标区域中显示网格

```
