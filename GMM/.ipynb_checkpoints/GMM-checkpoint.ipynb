```python
import cPickle as pickle
import numpy as np
import matplotlib.pyplot as plt
```

```python
colors = ['C0', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6']
```

```python
with open("cluster.pkl") as f:
    samples = pickle.load(f)
```

```python
for i in range(3):
    plt.scatter(samples[i][0], samples[i][1], c=colors[i], s=4)
    print samples[i][0].mean(), samples[i][1].mean(), samples[i][0].std(), samples[i][1].std()
plt.show()
```

```python
# np.random.randn生成的矩阵，里面的每个元素是独立取样的
# 一元正态分布，二元正态分布， 多元正态分布， 
# 多元独立正态 是 多元正态分布的特殊情形
mysamples = []
mu = [(-5,-3), (3,5), (10,8)]
sig = [(3,2), (2,1), (2,4)]
for i in range(3):
    x = sig[i][0] * np.random.randn(500) + mu[i][0]
    y = sig[i][1] * np.random.randn(500) + mu[i][1]
    mysamples.append((x,y))
```

```python
for i in range(3):
    plt.scatter(mysamples[i][0], mysamples[i][1], c=colors[i], s=4)
    print mysamples[i][0].mean(), mysamples[i][1].mean(), mysamples[i][0].std(), mysamples[i][1].std()
plt.show()
```

```python
def get_points(samples):
    num_per_cluster = []
    for i in range(3):
        num_per_cluster.append(samples[i][0].shape[0])

    points = []
    for i in range(3):
        points.append(np.concatenate([samples[i][0].reshape(num_per_cluster[i], -1), samples[i][1].reshape(num_per_cluster[i], -1)], axis=1))
    points = np.concatenate(points, axis=0)
    points = points.astype(np.float32)
    return points
```

```python
def compute_distmat(a, b):
    # 求 a 中每个点到 b 中每个点的 欧式距离
    # 其中：a \in m * d, b \in n * d
    # 返回 m * n 的矩阵
    M = np.dot(a, b.T)  # m * n
    te = np.square(a).sum(axis=1) # m * 1
    tr = np.square(b).sum(axis=1) # n * 1
    return np.array( -2*M + tr + np.matrix(te).T)
```

```python
# 输入：点 x，component数量
#     其中：x \in N * D，D表示维度，N表示点的数量
# 输出：pi, mu, sig, 以及每个点的属于每个component的概率
#     其中：pi \in K * 1, mu \in K * D, sig \in K * D * D, prob \in N * K
# 过程：
# 1. 初始化N_k, mu_k, sig_k
# 2. 估计每个点由每个component生成的概率 gamma_jk
# 3. 更新N_k, mu_k, sig_k    

def gmm(x, K):
    
    threshold = 1e-5
    N, D = x.shape
    idx = np.arange(N)
    np.random.shuffle(idx)
    centroids = x[idx[:3],:]

    def init():
        pi = np.array([0] * K, dtype=np.float32)
        mu = centroids
        sig = np.zeros((K,D,D), dtype=np.float32)

        distmat = compute_distmat(x, centroids)
        labels = distmat.argsort(axis=1)[:, 0]
        print labels

        for k in range(K):
            x_k = x[labels==k]
            pi[k] = x_k.shape[0]* 1. / N
            sig[k] = np.cov(x_k.T)

        print "init : ", pi.shape, mu.shape, sig.shape
        pgamma = np.zeros((N, K))
        pgamma[np.arange(N), labels] = 1
        return pi, mu, sig, pgamma
    
    def plot():
        assign_labels = pgamma.argsort(axis=1)[:, -1]
        for i in range(K):
            point_colors = [colors[i] for i in assign_labels]
            plt.scatter(x[:,0], x[:,1], c=point_colors, s=3)

        plt.show()

    pi, mu, sig, pgamma = init()
    plot()
#     return pi, mu, sig, pgamma

    def calc_prob():
        phi = []
        for k in range(K):
            xshift = x - mu[k]
            phi_k = np.exp(-0.5 * (xshift.dot(np.linalg.inv(sig[k])).dot(xshift.T)).diagonal())
            coef = (2*np.pi)**(D/2.) *  np.linalg.det(sig[k])**0.5        
            phi_k = phi_k / coef
            phi.append(phi_k.reshape(N,1))

        phi = np.concatenate(phi, axis=1)   # N * K
        pgamma = phi * pi.reshape(1,K)
        pgamma = pgamma / pgamma.sum(axis=1, keepdims=True)
        return pgamma, phi

    def update():
        pi = pgamma.sum(axis=0) / N
        mu = pgamma.T.dot(x) / (pi*N).reshape(K,1)
        for k in range(K):
            xshift = x - mu[k]
            sig[k] = xshift.T.dot(pgamma[:,k].reshape(N,1)*xshift)
            sig[k] /= (pi[k] * N)
        return pi, mu, sig

    Lprev = -np.inf
    while True:
        pgamma, phi = calc_prob()
        pi, mu, sig = update()
        
        L = np.sum(np.log((phi * pi.reshape(1, K)).sum(axis=1)))   # 似然函数，每个点由该 GMM 产生的概率
        print "likelihood : ", L, Lprev, L-Lprev
        
        if L - Lprev < threshold:
            break
        Lprev = L
        plot()

    return pi, mu, sig, pgamma
```

```python
k = 3
points = get_points(mysamples)
pi, mu, sig, pgamma = gmm(points, k)
```

```python
assign_labels = pgamma.argsort(axis=1)[:, -1]
```

```python
for i in range(k):
    point_colors = [colors[i] for i in assign_labels]
    plt.scatter(points[:,0], points[:,1], c=point_colors, s=3)

plt.show()
```
